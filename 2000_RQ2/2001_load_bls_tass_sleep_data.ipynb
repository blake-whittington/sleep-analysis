## Load Data

```python
bls = '../data_file/raw_data_csv/bls_sleep_data.csv'
bls_df = pd.read_csv(bls)
```

```python
bls_df = bls_df.drop(columns = ['Period', 'Label'])
bls_df = bls_df.rename(columns = {'Series ID': 'gender', 'Year':'year', 'Value':'avg_sleep'})
```

```python
bls_df.head(60)
```

```python
bls_df['gender'] = bls_df['gender'].replace({'TUU10101AA01000247':'total', 'TUU10101AA01000326': 'men', 'TUU10101AA01000405': 'women'})
```

```python
bls_df.head()
```

```python
total_df = bls_df[bls_df['gender'] == 'total']
total_df.head(3)
```

```python
men_df = bls_df[bls_df['gender'] == 'men']
men_df = men_df.reset_index(drop=True)
men_df.head(3)
```

```python
women_df = bls_df[bls_df['gender'] == 'women']
women_df = women_df.reset_index(drop=True)
women_df.head(3)
```

```python
bls_df.to_pickle('../data_file/cleaned_data_pickle/bls_df.pickle')
bls_df.to_csv(r'../data_file/cleaned_data_csv/bls_df.csv')
```

```python
total_df.to_pickle('../data_file/cleaned_data_pickle/total_df.pickle')
total_df.to_csv(r'../data_file/cleaned_data_csv/total_df.csv')
```
